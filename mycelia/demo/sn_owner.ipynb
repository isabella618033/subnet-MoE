```python
import requests

BASE_URL = "http://localhost:7000"


def get_phase():
    resp = requests.get(f"{BASE_URL}/get_phase")
    resp.raise_for_status()
    return resp.json()


def get_blocks_until_next_phase():
    resp = requests.get(f"{BASE_URL}/blocks_until_next_phase")
    resp.raise_for_status()
    return resp.json()


from mycelia.sn_owner.cycle import PhaseNames, PhaseResponse

PhaseResponse(**get_phase())
get_blocks_until_next_phase()[PhaseNames.train]
```

```python
from dataclasses import dataclass
from enum import Enum


@dataclass
class PhaseNames(str, Enum):
    distribute: str = "Distribute"  # miner download from validator
    train: str = "Train"  # miner trian
    commit: str = "Commit"  # miner commit hash and  vlaidators commit seed
    submission: str = "Submission"  # submit model
    validate: str = "Validate"  # validator validate
    merge: str = "Merge"  # validator merge


PhaseNames.merge == PhaseNames.merge
```
