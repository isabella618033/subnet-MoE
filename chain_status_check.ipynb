```python
def get_status(
        config: BaseConfig, 
        subtensor: bt.Subtensor
    ):
    all_commitments = subtensor.get_all_commitments(netuid = config.chain.netuid)
    metagraph = subtensor.metagraph(netuid = config.chain.netuid) 
    parsed: Dict[str, WorkerStatus] = {}

    print(all_commitments)
    for hotkey, commit in all_commitments.items():
        uid = metagraph.hotkeys.index(hotkey) 
        status_dict = json.loads(commit)
        try:
            if 'expert_group' in status_dict: # TODO: change it to using the neuron for classification 
                parsed[hotkey] = {'status': ValidatorStatus.model_validate(status_dict), 'neuron': metagraph.neurons[uid]} 
            else:
                parsed[hotkey] = {'status': MinerStatus.model_validate(status_dict), 'neuron': metagraph.neurons[uid]}
        except:
            pass

    return parsed

# status = get_status(config, sub)

status
```
